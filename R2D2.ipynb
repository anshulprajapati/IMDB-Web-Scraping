```python
imdb=[]
```

```python
for index in range(0, len(movies)):
    # Seperate movie into: 'place', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(0)
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],
            "link": links[index]}
    imdb.append(data)
```

```python
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=101&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=201&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=301&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=401&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=501&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=601&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=701&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=801&ref_=adv_nxt')
extraction_urls.append('https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=901&ref_=adv_nxt')
```

```python
for i in extraction_urls:
    url=i
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    movies_containers=soup.find_all('div',class_='lister-item mode-advanced')
    for j in range(0,len(movies_containers)):
        movie_title.append(movies_containers[j].h3.find('a').text)
    for j in range(0,len(movies_containers)):
        movie_rating.append(str(movies_containers[j].find('strong'))[8:11])
    for j in range(0,len(movies_containers)):
        s=movies_containers[j].find_all('a', href=True)
        temp=[]
        temp.append(s[-5].contents)
        temp.append(s[-4].contents)
        temp.append(s[-3].contents)
        temp.append(s[-2].contents)
        temp.append(s[-1].contents)
        crew.append(temp)
    for j in range(0,len(movies_containers)):
        rank.append(movies_containers[j].find('span', class_='lister-item-index unbold text-primary').text.replace('.',''))
    for j in range(0,len(movies_containers)):
        movie_year.append((movies_containers[j].h3.find('span', class_='lister-item-year text-muted unbold').text).replace('(','').replace(')',''))
```
